## **1. Definição do problema**


### **Contexto**
Desde o início dos anos 2000, a comunidade internacional tem dedicado esforços e compromissos sistêmicos para reduzir a taxa de mortalidade infantil, sobretudo a de crianças abaixo de cinco anos de idade (*under five mortality rate* - U5MR).

Em 2000, líderes de 189 países se encontraram na sede das Nações Unidas em Nova York e aprovaram a Declaração do Milênio, um compromisso para trabalharem juntos na construção de um mundo mais seguro, mais próspero e mais justo. Esta Declaração consolidou os **Objetivos de Desenvolvimento do Milênio (ODM)**, com foco em oito áreas prioritárias globais até 2015, incluindo Reduzir a mortalidade infantil, com a meta específica de **reduzir em dois terços a mortalidade de crianças menores que 5 anos**.

Como resultado, a taxa global de mortalidade de menores de cinco anos (U5MR) foi reduzida para [menos da metade](https://www.un.org/millenniumgoals/2015_MDG_Report/pdf/MDG%202015%20rev%20(July%201).pdf), de 90 para 43 mortes para cada 1.000 nascidos vivos entre 1990 e 2015.

Apesar do evidente progresso, os resultados chamaram a atenção para um cenário ainda mais sistêmico de desigualdades: a taxa U5MR permanecia 2x mais alta para crianças vivendo em países sub-desenvolvidos, em comparação àquelas residentes nos países mais ricos.

Os **Objetivos de Desenvolvimento Sustentável (ODS)**, estabelecidos em 2015 pela Assembleia Geral das Nações Unidas e o Conselho Econômico das Nações Unidas, trazem novos esforços conjuntos para promover uma nova visão de desenvolvimento: mais sustentável, integrada e sem deixar ninguém para trás. Com seus 17 objetivos ambiciosos e mensuráveis através de metas e indicadores específicos, os ODS representam uma agenda compartilhada entre países, Nações Unidas, sociedade civil, setor privado e academia para seguir respondendo aos diversos desafios globais.

Entre os 17 ODS, o ODS 3 - Saúde e bem-estar reforça o compromisso com a prevenção da mortalidade infantil:



---



**Meta 3.2: Até 2030, acabar com as mortes evitáveis de recém-nascidos e crianças menores de 5 anos, com todos os países objetivando reduzir a mortalidade neonatal para pelo menos 12 por 1.000 nascidos vivos e a mortalidade de crianças menores de 5 anos para pelo menos 25 por 1.000 nascidos vivos.**


---




Entretanto, apenas do sucesso observado durante o período de implementação dos ODM, o progresso em reduzir o U5MR vem caminhando a passos lentos, em comparação ao período anterior. Entre 2000 e 2024, 174 milhões de crianças com até cinco anos de idade morreram, especialmente em razão de causas preveníveis.

Nesse cenário preocupante, e na reta final para alcançar a taxa global de U5MR de 25/1.000 até 2030, fica a pergunta:

> **É possível predizer se a taxa U5MR prevista na Meta 3.2 será alcançada globalmente até 2030?**


### **Objetivo do MVP**
Este MVP tem o objetivo de **construir e avaliar um modelo capaz de prever a taxa anual de mortalidade de menores de cinco anos (U5MR) utilizando informações históricas dos países**.

Após a validação do desempenho em dados históricos, o modelo será utilizado para **gerar projeções para o período de 2025 a 2030**.

Este processo de aprendizado supervisionado de Machine Learning inclui uma abordagem baseline com modelos candidatos e discussão sobre as limitações do modelo.

### **Tipo de problema**

O problema em tela se refere à previsão da taxa de mortalidade de menores de cinco anos utilizando informações históricas da própria série temporal.

Portanto, nos referimos a um problema do tipo  **Séries temporais / forecasting**.  A variável-alvo U5MR é observada pelo UNICEF anualmente para cada país ao longo de três décadas, o que caracteriza uma série temporal por natureza. Neste sentido, vale destacar que serão empregados esforços para evitar o embaralhamento durante o treino.


### **Premissas e hipóteses**


**Hipóteses iniciais:**

`H1:` Limitações socioeconômicas e contextuais regionais, como choques externos, conflitos, pandemias, colapsos dos sistemas de saúde ou novas políticas públicas, não são modeladas neste MVP, embora possam influenciar significativamente a evolução da U5MR e, consequentemente, limitar a precisão das projeções realizadas pelo modelo.

`H2:` O recorte temporal escolhido (2000-2024) é representativo e captura os principais padrões de queda da mortalidade infantil, incluindo a influência dos compromissos globais assumidos entre 2000 e 2015 (ODM) e entre 2015 e 2030 (ODS).

`H3:` Trabalhar apenas com os valores medianos (Median) projetados pelo UNICEF, descartando os intervalos de incerteza Lower e Upper, é uma simplificação aceitável para o escopo deste MVP e não trará problemas sistêmicos para a construção e avaliação do modelo, tampouco para a interpretação das predições.

`H4:` A U5MR de cada país segue uma trajetória com estrutura suficiente para que modelos de forecasting capturem padrões além de uma simples reta.


### **Critérios de sucesso**


1. **Métrica principal**: Serão consideradas quatro métricas na eleição do modelo de machine learning mais adequado para responder ao problema de pesquisa:

* MAPE: permite a comparabilidade entre países;
* MAE: permite analisar o erro médio em mortes por 1.000 nascidos vivos;
* RMSE: permite analisar a sensibilidade a grandes erros;
* R²: permite analisar a capacidade explicativa do modelo.


O MAPE (Mean Absolute Percentage Error) será usado como métrica primária por permitir comparar o erro entre países em escalas muito diferentes. Por exemplo, um país com U5MR de 5 e outro com 120 não podem ser avaliados pelo mesmo erro absoluto. MAE, RMSE e R² serão considerados como métricas secundárias para complementar a análise.


2. **Resultados mínimos esperados:** O modelo será considerado satisfatório caso apresente desempenho superior ao modelo baseline no MAPE (obrigatório) e na maioria das outras métricas (complementar), mantendo R² superior a 0,80.

## **Funções auxiliares**

````python
def evaluate_regression(y_true, y_pred):
    """Calcula métricas básicas para regressão."""
    mse = mean_squared_error(y_true, y_pred)
    # Adicionando mean_absolute_percentage_error
    mape = mean_absolute_percentage_error(y_true, y_pred)
    return {
        "MAPE": mape,
        "MAE": mean_absolute_error(y_true, y_pred),
        "RMSE": np.sqrt(mse),
        "R2": r2_score(y_true, y_pred),

    }

def train_and_evaluate_model(model_name, pipeline, X_train, y_train, X_test, y_test):
    """
    Treina um modelo, realiza previsões, calcula as métricas
    e retorna os resultados.
    """
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    metrics = evaluate_regression(y_test, y_pred)
    print("=" * 60)
    print(f"MÉTRICA DE AVALIAÇÃO - {model_name}")
    print("=" * 60)
    display(pd.DataFrame([metrics]).T)
    return metrics

def evaluate_baseline(model_name, y_true, y_pred):
    """
    Avalia um modelo baseline e retorna suas métricas.
    """
    metrics = evaluate_regression(y_true, y_pred)
    print("=" * 60)
    print(f"MÉTRICA DE AVALIAÇÃO - {model_name}")
    print("=" * 60)
    display(pd.DataFrame([metrics]).T)
    return metrics

def add_forecast_layout(
    forecast_start=2024,
    forecast_end=2030,
    target=25,
    annotation=True
):
    """
    Adiciona os elementos comuns aos gráficos de forecast.
    """
    ax = plt.gca()
    # Linha vertical indicando o início das projeções
    ax.axvline(
        x=forecast_start,
        color="black",
        linestyle=":",
        linewidth=2
    )
    # Faixa sombreada do período projetado
    ax.axvspan(
        forecast_start + 1,
        forecast_end,
        color="gray",
        alpha=0.12
    )
    # Linha da meta ODS
    ax.axhline(
        target,
        color="red",
        linestyle="--",
        linewidth=2,
        label=f"Meta ODS ({target}/1.000)"
    )
    if annotation:
        xmin, xmax = ax.get_xlim()
        ymin, ymax = ax.get_ylim()
        # posicionamento sempre dentro da área do gráfico
        x_text = forecast_start - 0.6      # à esquerda da linha
        y_text = ymin + 0.92 * (ymax - ymin)
        y_arrow = ymin + 0.86 * (ymax - ymin)
        ax.annotate(
            "Início do Forecast",
            xy=(forecast_start, y_arrow),
            xytext=(x_text, y_text),
            ha="right",
            va="center",
            fontsize=10,
            fontweight="bold",
            arrowprops=dict(
                arrowstyle="-|>",
                color="black",
                lw=1.4,
                shrinkA=0,
                shrinkB=0
            )
        )
    ax.grid(
        axis="y",
        linestyle="--",
        alpha=0.4
    )

def prepare_forecast_plot(
    historical_df,
    forecast_df,
    group_columns,
    value_column="Under_Five_Mortality_Rate"
):
    """
    Prepara uma base única para gráficos contendo histórico e projeções.
    """
    historical = (
        historical_df
        .groupby(group_columns, as_index=False)[value_column]
        .mean()
    )
    historical["Tipo"] = "Histórico"

    forecast = (
        forecast_df
        .groupby(group_columns, as_index=False)[value_column]
        .mean()
    )
    forecast["Tipo"] = "Projeção"

    return pd.concat(
        [historical, forecast],
        ignore_index=True
    )


def get_top_countries(
    df,
    year=2024,
    top_n=5
):
    return (
        df
        .query("Year == @year")
        .nlargest(
            top_n,
            "Under_Five_Mortality_Rate"
        )["Country.Name"]
        .tolist()
    )

def get_region_order(
    df,
    reference_year=2000
):
    return (
        df
        .query("Year == @reference_year")
        .groupby("SDG.Region")[
            "Under_Five_Mortality_Rate"
        ]
        .mean()
        .sort_values(ascending=False)
        .index
        .tolist()
    )

def reorder_region_legend(
    hue_order,
    target_label="Meta ODS (25/1.000)"
):
    """
    Reorganiza automaticamente a legenda das regiões,
    mantendo a Meta ODS como último item.
    """

    ax = plt.gca()
    handles, labels = ax.get_legend_handles_labels()
    target_handle = None
    region_handles = {}
    for h, l in zip(handles, labels):

        if l == target_label:
            target_handle = h
        else:
            region_handles[l] = h
    ordered_handles = [
        region_handles[r]
        for r in hue_order
        if r in region_handles
    ]
    ordered_labels = [
        r
        for r in hue_order
        if r in region_handles
    ]
    if target_handle is not None:
        ordered_handles.append(target_handle)
        ordered_labels.append(target_label)
    ax.legend(
        ordered_handles,
        ordered_labels,
        title="Região",
        bbox_to_anchor=(1.02,1),
        loc="upper left"
    )

def classify_country(row):

    # País já havia atingido a meta em 2024
    if row["U5MR_2024"] <= 25:
        return "Já havia atingido"

    # País alcança a meta até 2030
    elif row["U5MR_2030"] <= 25:
        return "Atingiu entre 2025–2030"

    # Continua acima da meta
    else:
        return "Permanece acima da meta"


# PALETA PADRÃO PARA AS REGIÕES SDG

REGION_PALETTE = {
    "Sub-Saharan Africa": "#ff7f0e",
    "Central and Southern Asia": "#1f77b4",
    "Northern Africa and Western Asia": "#8c564b",
    "Eastern and South-Eastern Asia": "#2ca02c",
    "Oceania": "#e377c2",
    "Latin America and the Caribbean": "#7f7f7f",
    "Europe and Northern America": "#9467bd"
}
````

## **Sobre a fonte dos dados**

O dataset ser analisado é o **Under five mortality rate - U5MR**, parte da base de dados [UNICEF - Under five mortality data](https://data.unicef.org/resources/dataset/under-five-mortality-data/). A taxa U5MR mede a probabilidade de uma criança morrer entre o nascimento e exatamente cinco anos de idade, expressa por 1.000 nascidos vivos.

Este dataset foi escolhido por sua consistência histórica, com poucas alterações metodológicas e dados de quase todos os países desde 1970. Desde 2004, os dados são rigorosamente coletados, analisados e publicados pelo Grupo Interagencial das Nações Unidas para a Estimativa da Mortalidade Infantil (UN IGME), criado com o objetivo de compartilhar dados sobre mortalidade infantil, aprimorar os métodos de estimativa dessa mortalidade, relatar o progresso em direção às metas de sobrevivência infantil e fortalecer a capacidade dos países de produzir estimativas oportunas e devidamente avaliadas sobre a mortalidade infantil.

O UN IGME é liderado pelo Fundo das Nações Unidas para a Infância (UNICEF) e conta com a Organização Mundial da Saúde (OMS), o Grupo do Banco Mundial e a Divisão de População do Departamento de Assuntos Econômicos e Sociais das Nações Unidas (UNDESA) como membros plenos.

O relatório "[Levels & Trends in Child Mortality - Report 2025](https://data.unicef.org/wp-content/uploads/2026/05/UNIGME-Child-Mortality-Report-2025.pdf?client_id=301419196.1781617719&session_id=213622238)" traz informações detalhadas do percurso metodológico deste indicador, a partir de uma metodologia triangulada de levantamento de dados e avaliação da qualidade da informação para garantir estimativas mais precisas e ajustadas ao panorama global.

O dataset original possui um total de **600 instâncias e 82 atributos**. É necessário fazer algumas observações importantes que precisarão ser tratadas para avançarmos com o modelo.

* Limite superior (*upper*): a taxa máxima plausível, indicando um cenário mais grave, frequentemente devido a limitações de dados, conflitos ou pobreza.

### **Engenharia de Features**
Serão criadas três camadas de features, que serão utilizados pelo algoritmo para fazer previsões.

* **Year**: Tendência temporal. Já está presente no dataset.

* **Lag (defasagem temporal)**: Uma feature de lag consiste em utilizar valores passados da própria variável-alvo como variáveis explicativas. Como não dispomos de variáveis explicativas futuras (como renda, cobertura vacinal ou gastos em saúde), a melhor fonte de informação é o próprio histórico da série temporal. Assim, as features de lag permitem que o modelo aprenda a relação entre o passado e o futuro da U5MR. Aqui, serão criadas **três lags**.

* **Rolling means (média móvel)**: Essa feature captura uma visão mais estável da tendência, resumindo vários valores passados em apenas uma variável. Essa feature geralmente atua como complemento das Lags porque reduz o efeito de flutuações de curto prazo e revela o comportamento médio da taxa no tempo.

* **Variável categórica: SDG Region**: Essa variável será incluída para representar características estruturais compartilhadas entre países pertencentes à mesma região geográfica utilizada pelas Nações Unidas, conforme observado durante a análise exploratória. Como se trata de uma variável categórica nominal, faremos a transformação por meio de One-Hot Encoding durante o pipeline de pré-processamento.

Mais adiante, quando tivermos selecionado o modelo candidato com melhor performance, faremos um **feature importance** para analisar a importância dessas variáveis no modelo.


#### *Defasagem temporal (Lags)*

````python
# Criar features de lag para 'Under_Five_Mortality_Rate'
df_ml['mortality_rate_lag_1'] = df_ml.groupby('Country.Name')['Under_Five_Mortality_Rate'].shift(1)
df_ml['mortality_rate_lag_2'] = df_ml.groupby('Country.Name')['Under_Five_Mortality_Rate'].shift(2)
df_ml['mortality_rate_lag_3'] = df_ml.groupby('Country.Name')['Under_Five_Mortality_Rate'].shift(3)

# Preencher valores NaN resultantes dos lags (para os primeiros anos de cada país)
print('=' * 30)
print("DATAFRAME COM FEATURES DE LAG:")
print('=' * 30)
display(df_ml.head())
````

#### *Média móvel (Rolling Mean)*

````python
# Criar feature de Rolling Mean (média móvel) para 'Under_Five_Mortality_Rate'
df_ml["mortality_rate_rolling_mean"] = (
    df_ml
    .groupby("Country.Name")["Under_Five_Mortality_Rate"]
    .transform(
        lambda x:
            x.shift(1)
             .rolling(3)
             .mean()
    )
)

print('=' * 70)
print("DATAFRAME COM FEATURE DE ROLLING MEAN, APÓS APLICADO O shift(1): ")
print('=' * 70)
display(df_ml.head())
````

## **Divisão em Treino e Teste**

---


Para essa divisão, optei por fazer um **Out-of-time validation**, ao invés do clássico TimeSeriesSplit.

o `TimeSeriesSplit` foi desenvolvido para uma única série temporal. No caso desse dataset, existe um *panel data* com várias séries temporais distintas, uma para cada país (Brasil 2000-2024, Nigéria 2000-2024, Japão 2000-2024, etc). Como o TimeSeriesSplit apenas divide as linhas, ele não entenderia que existem, na verdade, 190 séries temporais independentes. Como consequência, poderiam ocorrer folds contendo anos diferentes de vários países, de forma difícil de interpretar e, dependendo da ordenação do DataFrame, até misturar contextos que não representam uma validação temporal limpa.

Já o `Temporal Holdout` garante que todos os países utilizam exatamente o mesmo horizonte temporal.

Vale destacar que a lógica de 70/30, 80/20 ou 75/25 é uma convenção mmenos utilizada em séries temporais. Neste tipo de aprendizado, **preservar a ordem cronológica é mais importante do que respeitar uma proporção fixa**.

````python
# Remover linhas sem informações suficientes para treinamento
df_ml = (
    df_ml
    .dropna(
        subset=[
            "target_next_year",
            "mortality_rate_lag_1",
            "mortality_rate_lag_2",
            "mortality_rate_lag_3",
            "mortality_rate_rolling_mean"
        ]
    )
    .sort_values(
        ["Country.Name", "Year"]
    )
    .reset_index(drop=True)
)

# Definição da variável-alvo e das features

TARGET = "target_next_year"

FEATURES = [
    "Year",
    "mortality_rate_lag_1",
    "mortality_rate_lag_2",
    "mortality_rate_lag_3",
    "mortality_rate_rolling_mean",
    "SDG.Region"
]

X_full = df_ml[FEATURES]
y_full = df_ml[TARGET]


# Divisão temporal (Temporal Holdout)
TEST_YEAR_START = 2023
TRAIN_YEAR_END = TEST_YEAR_START - 1

# Máscaras para separar treino e teste
train_mask = X_full["Year"] <= TRAIN_YEAR_END
test_mask = X_full["Year"] >= TEST_YEAR_START

# Conjuntos de treino
X_train = X_full.loc[train_mask].copy()
y_train = y_full.loc[train_mask].copy()

# Conjuntos de teste
X_test = X_full.loc[test_mask].copy()
y_test = y_full.loc[test_mask].copy()

# DataFrames completos (úteis para EDA, forecast e gráficos)
df_train = (
    df_ml
    .loc[train_mask]
    .copy()
)

df_test = (
    df_ml
    .loc[test_mask]
    .copy()
)

# Resumo da divisão temporal
print("=" * 70)
print("DIVISÃO TEMPORAL DO CONJUNTO DE DADOS")
print("=" * 70)
print(f"Período de treinamento : {X_train['Year'].min()}–{X_train['Year'].max()}")
print(
    f"Período de teste       : "
    f"{X_test['Year'].min()}–{X_test['Year'].max()}"
)
print("\n")
print("Observação:")
print(
    "O ano de 2024 não integra os conjuntos de treinamento ou teste "
    "porque sua variável-alvo (U5MR de 2025) ainda não é observada. "
    "Seus valores são utilizados apenas como condição inicial do forecasting recursivo de 2025–2030."

)
print("\n")
print('=' * 70)
print("Dimensões dos conjuntos:")
print('=' * 70)
print(f"X_train: {X_train.shape}")
print(f"y_train: {y_train.shape}")
print(f"X_test : {X_test.shape}")
print(f"y_test : {y_test.shape}")
print("\n")
print('=' * 70)
print("Features utilizadas pelo modelo:")
print('=' * 70)
print(FEATURES)
print("\n")
print('=' * 70)
print("Primeiras observações do conjunto de treinamento:")
print('=' * 70)
display(X_train.head())
print("\n")
print('=' * 70)
print("Primeiros valores da variável-alvo:")
print('=' * 70)
display(y_train.head())
````

## **Pipeline de preparação dos dados**

````python
# Identificar as features numéricas e categóricas
# 'SDG.Region' agora é uma feature categórica a ser encodada.
numerical_features = [
    'Year',
    'mortality_rate_lag_1',
    'mortality_rate_lag_2',
    'mortality_rate_lag_3',
    'mortality_rate_rolling_mean'
]
categorical_features = ['SDG.Region']

# Criar o pré-processador
# Aplica SimpleImputer e StandardScaler a features numéricas e OneHotEncoder a features categóricas.
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='mean')), # Adicionado SimpleImputer
            ('scaler', StandardScaler())
        ]), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Criar o pipeline para Linear Regression
linear_regression_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression()) # Alterado de volta para LinearRegression
])


# Criar o pipeline para Random Forest Regressor
random_forest_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(random_state=SEED)) # RandomForestRegressor
])

# Criar o pipeline para Gradient Boosting Regressor
gradient_boosting_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', GradientBoostingRegressor(random_state=SEED)) # GradientBoostingRegressor
])
````

## **Otimização de hiperparâmetros**

````python
param_dist = {
    'regressor__max_depth': randint(3, 11),
    'regressor__n_estimators': randint(100, 500),
    'regressor__min_samples_leaf': randint(1, 20),
    'regressor__max_features': ['sqrt', 'log2', 0.8, 1.0]
}

# Usaremos o pipeline de Random Forest que já inclui o pré-processador
random_forest_pipeline_tuned = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(random_state=SEED))
])

# Configurar RandomizedSearchCV
random_search_rf = RandomizedSearchCV(
    estimator=random_forest_pipeline_tuned,
    param_distributions=param_dist,
    n_iter=15,  # Número de combinações de parâmetros a serem amostradas
    scoring='neg_mean_absolute_error', # Otimizar para MAE (negativo, pois GridSearchCV maximiza)
    cv=3, # 3-fold cross-validation
    verbose=1,
    n_jobs=-1, # Usar todos os cores disponíveis
    random_state=SEED
)

random_search_rf.fit(X_train, y_train)

print("\n")
print('=' * 50)
print("MELHORES HIPERPARÂMETROS ENCONTRADOS:")
print('=' * 50)
print(random_search_rf.best_params_)
print("\n")
print('=' * 50)
print("MELHOR MAE NA VALIDAÇÃO CRUZADA:")
print('=' * 50)
print(-random_search_rf.best_score_)

# Atribuir o melhor modelo encontrado
best_random_forest_pipeline = random_search_rf.best_estimator_

# Avaliar o melhor modelo no conjunto de teste
y_pred_rf_tuned = best_random_forest_pipeline.predict(X_test)
metrics_rf_tuned = evaluate_regression(y_test, y_pred_rf_tuned)

print("\n")
print('=' * 80)
print("MÉTRICAS DE AVALIAÇÃO DO RANDOM FOREST OTIMIZADO NO CONJUNTO DE TESTE::")
print('=' * 80)
display(pd.DataFrame([metrics_rf_tuned]).T)

# Atualizar os resultados gerais com o modelo otimizado
all_model_results["Random Forest (Tuned)"] = metrics_rf_tuned

print("\n")
print('=' * 80)
print("COMPARAÇÃO FINAL DE TODOS OS MODELOS (incluindo Random Forest Otimizado):")
print('=' * 80)
display(pd.DataFrame(all_model_results).T)
````

Assim, sua escolha como modelo final mostra-se consistente tanto sob a perspectiva das métricas agregadas quanto na análise individual de países com diferentes perfis epidemiológicos.


## **Forecast recursivo: 2025-2030**
A sub-seção anterior alcançou, com sucesso, o primeiro objetivo deste MVP: construir e avaliar um modelo capaz de prever a taxa anual de mortalidade de menores de cinco anos (U5MR) utilizando informações históricas dos países.

A seguir, o modelo será novamente testado para que se alcance o segundo objetivo do projeto de aprendizado supervisionado: gerar projeções para o período de 2025 a 2030.

````python
# Modelo: Random Forest Otimizado
forecast_model = best_random_forest_pipeline

print("=" * 70)
print("FORECAST RECURSIVO DA U5MR")
print("=" * 70)
print(f"Modelo utilizado : {forecast_model.named_steps['regressor'].__class__.__name__}")

# Informações iniciais
last_historical_year = df_mortality_tidy["Year"].max()
years_to_forecast = list(
    range(last_historical_year + 1,
          last_historical_year + 7)
)

countries_to_forecast = sorted(
    df_mortality_tidy["Country.Name"].unique()
)

print(f"Último ano observado : {last_historical_year}")
print(f"Período projetado    : {years_to_forecast[0]}–{years_to_forecast[-1]}")
print(f"Número de países     : {len(countries_to_forecast)}")


# Mapeamento País → Região
country_region_map = (
    df_mortality_tidy
    .drop_duplicates("Country.Name")
    .set_index("Country.Name")["SDG.Region"]
    .to_dict()
)

# Mapeamento País → ISO
country_iso_map = (
    df_mortality_tidy
    .drop_duplicates("Country.Name")
    .set_index("Country.Name")["ISO.Code"]
    .to_dict()
)

# Histórico dos países
country_histories = {
    country: history.sort_values("Year")
    for country, history in
    df_mortality_tidy.groupby("Country.Name")
}

# Inicializar lags
current_values = {}

for country, history in country_histories.items():

    last_values = (
        history
        .tail(3)["Under_Five_Mortality_Rate"]
        .to_numpy()
    )

    current_values[country] = {

        "lag_1": last_values[-1],   # 2024
        "lag_2": last_values[-2],   # 2023
        "lag_3": last_values[-3]    # 2022

    }

# Forecast recursivo
forecast_records = []

for year in years_to_forecast:

    rows = []

    for country in countries_to_forecast:

        values = current_values[country]

        rolling_mean = np.mean([
            values["lag_1"],
            values["lag_2"],
            values["lag_3"]
        ])

        rows.append({

            "Year": year,
            "mortality_rate_lag_1": values["lag_1"],
            "mortality_rate_lag_2": values["lag_2"],
            "mortality_rate_lag_3": values["lag_3"],
            "mortality_rate_rolling_mean": rolling_mean,
            "SDG.Region": country_region_map[country],
            "Country.Name": country

        })


    # Previsão para todos os países do ano corrente
    df_features = pd.DataFrame(rows)

    X_predict = df_features.drop(columns="Country.Name")

    predictions = forecast_model.predict(X_predict)

    # Atualizar lags (forecast recursivo)
    for country, prediction in zip(countries_to_forecast,
                                   predictions):
        prediction = np.clip(prediction,
                             a_min=0,
                             a_max=None)
        forecast_records.append({
            "Country.Name": country,
            "ISO.Code": country_iso_map[country],
            "Year": year,
            "Under_Five_Mortality_Rate": prediction,
            "Type": "Predicted"

})

        values = current_values[country]
        values["lag_3"] = values["lag_2"]
        values["lag_2"] = values["lag_1"]
        values["lag_1"] = prediction

# Resultado final
df_forecast_results = (
    pd.DataFrame(forecast_records)
)

df_forecast_results["Under_Five_Mortality_Rate"] = (
    df_forecast_results["Under_Five_Mortality_Rate"]
    .round(2)
)

print("\nForecast concluído com sucesso!")

display(df_forecast_results.head())
display(df_forecast_results.tail())
````

Com base na saída apresentada, o forecast foi executado conforme o esperado. A saída acima confirma que todas as etapas previstas na modelagem foram concluídas corretamente.

Como podemos inferir pelas impressões da cabeça e da cauda do dataset, as previsões parecem plausíveis: todos permanecem em faixas compatíveis com seus perfis históricos e sem valores impossíveis, como mortalidade negativa, centenas de mortes por mil ou saltos abruptos logo no primeiro ano. Portanto, **podemos afirmar que o modelo de machine learning desenvolvido para gerar projeções para o período de 2025 a 2030 funcionou de forma apropriada e sem anomalias, cumprindo o objetivo do MVP**.

O modelo selecionado produziu projeções coerentes com o comportamento histórico da mortalidade infantil e apresentou desempenho superior ao benchmark adotado. Além disso, o forecast permitiu identificar quais países tendem a permanecer acima da meta dos Objetivos de Desenvolvimento Sustentável até 2030 e quais possuem maior probabilidade de alcançá-la durante o período projetado.


## **Geração de valor**

Embora desenvolvido como um MVP relativamente simples de aprendizado supervisionado de machine learning, combinando boas práticas de engenharia e análise de dados, o produto apresenta potencial de aplicação prática para organizações e tomadores de decisão que atuam com monitoramento de indicadores sociais, especialmente organismos internacionais, governos nacionais e instituições de pesquisa.

Uma das principais contribuições do modelo consiste em transformar séries históricas em projeções quantitativas capazes de antecipar cenários futuros. Em vez de avaliar apenas indicadores retrospectivos, gestores podem utilizar previsões para identificar países ou regiões com maior risco de não atingir metas internacionais (ou aqueles com maior probabilidade de alcançá-lo, considerando a reta final de cinco anos), permitindo priorizar investimentos, direcionar ações de cooperação técnica e fortalecer estratégias de prevenção da mortalidade infantil.

O forecast também fornece uma visão prospectiva do progresso esperado até o horizonte dos Objetivos de Desenvolvimento Sustentável, funcionando como instrumento complementar aos relatórios anuais tradicionalmente publicados pelas Nações Unidas. Ainda que essas projeções não substituam análises epidemiológicas completas, elas podem servir como mecanismo de apoio à tomada de decisão baseada em evidências.


## **Limitações**

Apesar dos resultados positivos, este MVP apresenta algumas limitações importantes. Estas limitações podem ser consideradas como previsíveis, considerando o conhecimento ainda em crescimento do autor no campo da ciência de dados.

A primeira diz respeito ao **conjunto de variáveis utilizadas**. O modelo foi construído exclusivamente a partir do histórico da própria U5MR e da região geográfica, não incorporando determinantes socioeconômicos relevantes, como renda, cobertura vacinal, gastos públicos em saúde, conflitos armados, eventos climáticos extremos ou indicadores demográficos. Dessa forma, choques estruturais futuros permanecem fora da capacidade explicativa do modelo.

Como consequência dessa limitação, observou-se a **não confirmação parcial da hipótese H1**. A análise exploratória indicava que a região geográfica provavelmente desempenharia papel importante na explicação da mortalidade infantil. Entretanto, a análise de importância das variáveis mostrou que, para o Random Forest, praticamente toda a capacidade preditiva concentrou-se na própria dinâmica temporal da série, enquanto a variável regional apresentou contribuição bastante reduzida. Isso não significa que fatores regionais sejam irrelevantes para a mortalidade infantil, mas sim que, dentro da estrutura deste modelo e utilizando apenas as variáveis disponíveis, essas diferenças já estavam amplamente incorporadas pelo histórico recente de cada país.

Outra limitação decorre da própria **natureza do forecast recursivo**. Como cada previsão é utilizada para gerar a previsão do ano seguinte, pequenos erros podem se propagar ao longo do horizonte temporal. Embora essa propagação não tenha produzido instabilidades relevantes até 2030, horizontes mais longos provavelmente exigiriam abordagens específicas para reduzir o acúmulo de erros.




## **Possíveis próximos passos**

Como evolução natural deste trabalho, diversas melhorias poderiam ser incorporadas em versões futuras.

Entre elas destacam-se:

* inclusão de variáveis socioeconômicas (como PIB per capita e Índice de Desenvolvimento Humano), além de dados sobre a prevalência das principais doenças preveníveis associadas à mortalidade infantil nesta faixa etária: pneumonia, diarréia, malária, complicações neonatais e asfixia durante o parto. Essas outros features certamente trariam maior riqueza ao desenvolvimento do algoritmo, levando a conclusões mais precisas e adaptadas à realidade.
* comparação com outros algoritmos específicos para séries temporais, como XGBoost para dados temporais, LightGBM, CatBoost, Prophet e modelos baseados em LSTM ou Transformers;
* utilização de estratégias mais robustas de validação temporal, como *rolling-origin evaluation*;
* construção de intervalos de previsão, quantificando explicitamente a incerteza associada às projeções;
* desenvolvimento de um dashboard para acompanhamento contínuo das projeções por país, região e meta dos ODS.
